## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

## Constants

```python
c = 299792458      #speed of light, m/s
hc = 1239.84193    #Plank's constant*c, eV*nm
ea_s = []          #TODO
err_ea_s = []      #TODO

xl_path = '/Users/xnimir/Desktop/EA Cs/Spread sheets/Sets/'      #directory for reading/writing data from/to
xl_name = 'Cs_EA_set4_Feb19.xlsx'                                #excel file to read
save_img = False   #save image?
file_name = ''     #file name for the saved plot (png, dpi = 300)

xl_sheets = ['Co1','Counter1','Co2','Counter2','Co3'] #excel sheet names from file
columns = ['Signal',' WL red (nm)','Ion current (pA)','Laser power red (mW)','std dev','pow dev'] #excel column names
```

## Python list to NP array

```python
def np_array(l):
    return np.array(l.to_numpy().flatten())
```

## Create x, y, and error bars

```python
def create_plot_data(sheet):
    d_params = pd.read_excel(xl_path + xl_name, sheet_name = sheet) #reading excel sheet into data frame
    data = []                                                       #list of excel data from sheet
    for col in columns:                                             #building list by looping over each column 
        df_s = pd.DataFrame(d_params, columns = [col])              #creating data frame given d_params for col
        data.append(np_array(df_s))                                 #appending numpy array to list
   
    for j, counts in enumerate(data[0]):          #data[0] = signal, j = row index, counts = signal counts 
        if counts == 0:                           #deleting any zeros in sheet
            for i in range(len(data)):            #i = column, looping over columns in data
                np.delete(data[i], j)             #deleting row j in column i of data
   
    ens = hc / data[1]                            #converting WL from nm to eV
    s_n = data[0] / (data[2] * data[3])           #normalizing signal to ion current and laser power 
    

    #this is an error propagation which accounts for signal, ion current, and laser power
    #df_ds,c,p = (partial derivative with respect to... * deviation)^2 
    df_ds = (np.sqrt(data[0]) / (data[2] * data[3]))**2       #...signal (poisson dist. )
    df_dc = (data[0] / (data[2]**2 * data[3]) * data[4])**2   #...current
    df_dp = (data[0] / (data[2] * data[3]**2) * data[5])**2   #...power
    
    var_s = np.sqrt(df_ds + df_dc + df_dp) #std dev 
    
    x1 = ens[:]            #array of x values in eV                
    y1 = s_n[:]            #array of normalized y values
    err_y = var_s[:]       #array of error bars 
    
    return(x1, y1, err_y)
```

## Wigner law fit

```python
def Wigner(x, a, b, ea):
    return a + b * np.sqrt(x - ea + np.abs(x - ea))
```

## Guess parameters for Wigner law fit

```python
def guess_fit_pars(sheet):
    #guess parameters for Wigner fit in co progagation scheme (this is the default)
    pars = [1.7, 300, 1.9268]      #[a, b, ea] 
    
    #guess parameters for Wigner fit in counter propagation scheme
    if "counter" in sheet.lower(): 
        pars = [1, 100, 1.92563]   #[a, b, ea] 
        
    xarray = np.arange(pars[2] - 100e-6, pars[2] + 200e-6, 1e-7)        #returns evenly spaced values for the x-axis 
    y_func = Wigner(xarray, *pars)                                      #definition of the trial Wigner curve
    bound_par = ([0, 0, pars[2] - 0.0002], [5, 5000, pars[2] + 0.0001]) #bounds for fitting guess 
    
    return(pars, xarray, y_func, bound_par)
```

## Plot a sheet (eV) 

```python
def plot_sheet(k, sheet):
    x1, y1, err_y = create_plot_data(sheet)
    pars, xarray, y_func, bound_par = guess_fit_pars(sheet)
    
    try:
        pl0, covl0 = curve_fit(Wigner, x1, y1, [*pars], bounds = bound_par, sigma = err_y, absolute_sigma = True, maxfev = 1000)
        plt.figure(k)
        plt.errorbar(x1, y1, yerr = err_y, fmt = '.k', color = 'blue', capthick = 0.5, capsize = 5, elinewidth = 0.5)
        plt.plot(xarray, Wigner(xarray, *pl0), label = 'fit', color = 'black')
        #plt.plot(xarray, y_func, linestyle = '--', alpha = 0.6) #plotting of the trial Wigner curve
        plt.scatter(x1, y1, s = 7)
        plt.xlabel("Photon energy (eV)")
        plt.ylabel("Counts (normalized)")
        st_error = np.sqrt(np.diag(covl0))[2] #takes cov Matrix and gives the std of the parameters
        err_ea_s.append(st_error)
        plt.title(sheet)
        middle_trans = 1.454620680 # energy diff between Ground state and 6p in eV
        #Doppler = 600*1e-6 # this is just to estimate the Doppler shift if needed
        electron_affinity = abs(pl0[2] - middle_trans)
        ea_s.append(electron_affinity)
        plt.text(x1[0], y1[7], 'EA: ' + str(round(electron_affinity,7)) + ' eV', bbox = dict(facecolor = 'red', alpha = 0.3))
        print (round(electron_affinity,7), 'eV', 'statistichal error:', str(round(st_error,7)))
        
        if save_img:
            plt.savefig(_path + file_name + '_{}.png'.format(sheet), dpi = 300, bbox_inches = 'tight')
        plt.show()
        
        return(pl0)
    
    except Exception as e:
        print (e)
        
    
plot_sheet(0, xl_sheets[0])
```

## Plot multiple sheets (eV)

```python
def plot_sheets():
    for k, sheet in enumerate(xl_sheets):
        plot_sheet(k, sheet)    

plot_sheets()
```

## Plot with wavelength (nm) on x-axis

```python
def plot_sheets_wl(k, sheet): 
    x1, y1, err_y = create_plot_data(sheet)
    pl0 = plot_sheet(k, sheet)
    
    wls = hc / x1
    wl_ea = hc / pl0[2] # gives the threshold wavelength
    plt.figure(k + 20)
    plt.scatter(wls, y1, s = 5)
    plt.errorbar(wls, y1, yerr = err_y, fmt = '.k', color = 'blue', capthick = 0.5, capsize = 5, elinewidth = 0.5)
    plt.axvline(x = wl_ea, color = 'pink', linestyle = '--')
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Counts (normalized)")
    plt.xlim([wl_ea - 0.03, wl_ea + 0.07])
    plt.text(wls[2], y1[3], xl_sheets[k] + ','+ str(round(wl_ea,4)) + ' nm', bbox = dict(facecolor = 'red', alpha = 0.3))

    if save_img:
        plt.savefig(xl_path + file_name + 'wl_{}.png'.format(sheet), dpi = 300, bbox_inches = 'tight')
    plt.show() 

plot_sheets_wl(0, xl_sheets[0])
```
